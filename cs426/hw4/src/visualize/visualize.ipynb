```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def readResult(K):
    with open("../result/K-%d.out" % K, "r") as f:
        lines = f.readlines()
        clusters = [[float(xi) for xi in line.split()] for line in lines[:K]]
        c_idx = [int(c) for c in lines[K].split()]
        
    return clusters, c_idx
def readPoints():
    with open("../data/data.csv", "r") as f:
        lines = f.readlines()[1:]
        points = [[float(xi) for xi in line.split(',')[:3]] for line in lines]
        label = [int(float(line.split(',')[-1])) for line in lines]
    return points, label
```

```python
c2, c2_idx = readResult(2)
c3, c3_idx = readResult(3)
c4, c4_idx = readResult(4)
```

```python
print(c2), print(c2_idx)
```

```python
points, label = readPoints()
```

```python
import pylab as p
import mpl_toolkits.mplot3d.axes3d as p3
color = 'bgrc'
fig = None
def plotCluster(points, idx):
    global fig
    fig=p.figure()
    ax = p3.Axes3D(fig)
    x, y, z = list(zip(*points))
    c = [color[i] for i in idx]
    ax.scatter(x, y, z, c=c)
```

```python
plotCluster(points, c2_idx)
```

```python
fig.savefig("k2.eps", dpi=1200)
```

```python
c3, c3_idx = readResult(3)
plotCluster(points, c3_idx)
```
