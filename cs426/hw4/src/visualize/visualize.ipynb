```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def readResult(K):
    with open("../result/K-%d.out" % K, "r") as f:
        lines = f.readlines()
        clusters = [[float(xi) for xi in line.split()] for line in lines[:K]]
        c_idx = [int(c) for c in lines[K].split()]
        
    return clusters, c_idx
def readPoints():
    with open("../data/data.csv", "r") as f:
        lines = f.readlines()[1:]
        points = [[float(xi) for xi in line.split(',')[:3]] for line in lines]
        label = [int(float(line.split(',')[-1])) for line in lines]
    return points, label
points, label = readPoints()
```

```python
points, label = readPoints()
```

```python
import pylab as p
import mpl_toolkits.mplot3d.axes3d as p3
color = 'bgrcmykw'
fig = None
def plotCluster(points, idx):
    global fig
    fig=p.figure()
    ax = p3.Axes3D(fig)
    x, y, z = list(zip(*points))
    c = [color[i] for i in idx]
    ax.scatter(x, y, z, c=c)
    
def plotClusterLine(points, idx, clusters):
    fig, axl = plt.subplots(subplot_kw={'projection': '3d'})
    for i, point in enumerate(points):
        x, y, z = list(zip(*[point, clusters[idx[i]]]))
        axl.plot(x, y, z, color=color[idx[i]])

def dist(p1, p2):
    d = 0.0
    for i in range(len(p1)):
        d += (p1[i] - p2[i]) ** 2
    return d
```

```python
c2, c2_idx = readResult(2)
plotCluster(points, c2_idx)
plotClusterLine(points, c2_idx, c2)
```

```python
fig.savefig("k2.eps", dpi=1200)
plt.savefig("k2line.eps", dpi=1200)
```
